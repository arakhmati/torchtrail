```python
import torch
import torchtrail
```

```python
with torchtrail.trace():
    tensor = torch.rand(1, 64)
    tensors = torch.split(tensor, split_size_or_sections=32, dim=1)
    tensor = torch.cat(tensors, dim=1)
torchtrail.visualize(tensor)
```

```python
torchtrail.visualize(tensor, verbose=True)
```
