```python
import torch
import torchtrail
```

```python
with torchtrail.trace():
    tensor = torch.rand(8, 128, 12, 64)
    tensor = torch.permute(tensor, (0, 2, 1, 3))
torchtrail.visualize(tensor)
```

```python
torchtrail.visualize(tensor, verbose=True)
```
